# Problem Statement

    In the telecom industry, customers are able to choose from multiple service providers and actively switch from one operator to another. In this highly competitive market, the telecommunications industry experiences an average of 15-25% annual churn rate. Given the fact that it costs 5-10 times more to acquire a new customer than to retain an existing one, customer retention has now become even more important than customer acquisition.

    For many incumbent operators, retaining high profitable customers is the number one business goal. To reduce customer churn, telecom companies need to predict which customers are at high risk of churn. In this project, you will analyze customer-level data of a leading telecom firm, build predictive models to identify customers at high risk of churn.

    In this competition, your goal is to build a machine learning model that is able to predict churning customers based on the features provided for their usage.

```python
#All numerical variables in training data are on numtrain
num_train = tltrain_df.columns[(tltrain_df.dtypes=="float64")|(tltrain_df.dtypes=="int64")].tolist()
num_train.remove('churn_probability')
print('Training dataset Numeric Column count: ', len(num_train))

#All numerical variables in training data are on numtrain
num_test = tltest_df.columns[(tltest_df.dtypes=="float64")|(tltest_df.dtypes=="int64")].tolist()
print('Testing dataset Numeric Column count: ' , len(num_test))


#All categorical var in training data are in cat_train and removing target var
cat_train = tltrain_df.select_dtypes(exclude=np.number).columns.tolist()
print('Training dataset Non-Numeric Column count: ' , len(cat_train))

#All categorical var in training data are in cat_train and removing target var
cat_test = tltest_df.select_dtypes(exclude=np.number).columns.tolist()
print('Testing dataset Numeric Column count: ' , len(cat_test))
```

So in the above section both train and test data are catagorised to numerical and categorical data based all float and int data type variables are considered as numerical and object type considered as categorical data.

```python
# Printing columns with above 99% of unique values as well as single unqiue value
for column in tltest_df.columns:
#     getting the count of unique values
    unique_count = len(tltrain_df[column].unique())
#     checking for unique value count more than 99% as well as single value
    if unique_count == 1 or ((unique_count/tltrain_df.shape[0])*100) >= 99 :
        print('{0:20}  {1}'.format(column, unique_count))
```
